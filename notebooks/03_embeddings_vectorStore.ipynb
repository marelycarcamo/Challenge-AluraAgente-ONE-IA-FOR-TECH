## ¿Qué es un embedding?

Un embedding es una representación numérica de un texto.

Los modelos de embeddings transforman palabras, frases o documentos en vectores de alta dimensión donde textos con significados similares quedan cercanos entre sí.

Esta representación permite que el sistema encuentre información por significado y no únicamente por coincidencia exacta de palabras.

## Modelo seleccionado

Para la versión 1.0 de SOPHIA se utilizará el modelo:

all-MiniLM-L6-v2

Motivos de la elección:

- Modelo liviano y rápido.
- Funciona correctamente en CPU.
- Buen equilibrio entre rendimiento y calidad.
- Ampliamente utilizado en proyectos RAG.
- Adecuado para un MVP orientado a portafolio.

### Importación de librerías

````python
# Esta librería sentence-transformes nos permite descargar el modelo, cargarlo y generar embeddings.
# %pip install sentence-transformers
````

### Carga del modelo

````python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")
````

### Generación de un embedding de prueba

Como primera validación se generará el embedding de un texto simple para comprobar que el modelo funciona correctamente.

````python
texto = """
La Municipalidad de Valdivia cuenta con un Plan Comunal de Emergencia
para coordinar la respuesta frente a desastres naturales.
"""
````

````python
embedding = model.encode(texto)
````

## Inspección del embedding

El resultado es un vector numérico que representa el significado semántico del texto.

No es necesario interpretar individualmente cada valor; lo relevante es que el modelo utiliza este vector para calcular similitudes entre textos.

````python
type(embedding)
````

````python
print("Dimensiones:", len(embedding))
````

````python
embedding[:10]
````

## Reflexión técnica

Durante esta etapa se comprobó que el modelo all-MiniLM-L6-v2 genera representaciones vectoriales de 384 dimensiones.

Estos vectores constituyen la base del mecanismo de búsqueda semántica que utilizará SOPHIA en la etapa de Retrieval.

La generación de embeddings separa el significado del texto de su representación original, permitiendo encontrar información relevante incluso cuando la consulta utiliza palabras distintas a las presentes en los documentos.

## Validación del modelo

Se comprobó que el modelo `all-MiniLM-L6-v2` genera un vector de **384 dimensiones** para cualquier texto de entrada.

Aunque la longitud del vector permanece constante, sus valores cambian según el contenido del texto. Esto confirma que el modelo representa el significado semántico de cada texto mediante una representación numérica de tamaño fijo.

Esta propiedad permitirá, en las siguientes etapas del proyecto, comparar consultas y documentos utilizando medidas de similitud semántica en lugar de coincidencias exactas de palabras.
